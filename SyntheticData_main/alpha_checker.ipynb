```python
import os
import numpy as np
from PIL import Image

def process_image(image_path):
    """
    Processes a single image to analyze its alpha channel.

    Args:
        image_path (str): The path to the image file.
    """
    with Image.open(image_path) as img:
        if img.mode in ('RGBA', 'LA') or ('transparency' in img.info):
            print(f"\nProcessing image: {image_path}")
            img_arr = np.array(img)
            if img_arr.ndim == 3 and img_arr.shape[2] >= 4:
                alpha = img_arr[:, :, 3]  # Extract the alpha channel
            else:
                print("Alpha channel not found in the image.")
                return

            total_pixels = alpha.size
            fully_opaque = np.sum(alpha == 255)  # Count fully opaque pixels
            non_zero = np.sum(alpha > 0)  # Count pixels with non-zero alpha
            print(f"Total alpha channel pixels: {total_pixels}")
            print(f"Fully opaque pixels (alpha == 255): {fully_opaque}")
            print(f"Non-zero alpha pixels (alpha > 0): {non_zero}")
        else:
            print(f"\nImage {image_path} does not contain an alpha channel.")

def process_directory(directory):
    """
    Processes all PNG images within a given directory and its subdirectories.

    Args:
        directory (str): The path to the directory containing images.
    """
    for root, _, files in os.walk(directory):
        for file in files:
            if file.lower().endswith('.png'):
                image_path = os.path.join(root, file)
                process_image(image_path)

if __name__ == '__main__':
    directory = r"C:\Users\Admin\PycharmProjects\PythonProject\Human-Segmentation-main\SyntheticData_main\Obj"
    # directory = r"C:\Users\Admin\PycharmProjects\PythonProject\Human-Segmentation-main\SyntheticData_main\Result\epoch19\label"
    process_directory(directory)
```
